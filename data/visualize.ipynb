```python
import pandas as pd
import matplotlib.pyplot as plt

t = pd.read_csv("~/SageMaker/data/tweets/labelled.csv", sep = ';')
t["created_at"] = t["created_at"].apply(pd.to_datetime)
t["created_at"] = t["created_at"].dt.date

fig, axes = plt.subplots(nrows=1, ncols=1)
t['created_at'].hist(ax=axes, xrot = 45)

# set title and axis labels
plt.suptitle('Tweet Data', x=0.5, y=1.05, ha='center', fontsize='xx-large')
fig.text(0.5, -0.15, 'Date', ha='center')
fig.text(-0.02, 0.5, 'Num', va='center', rotation='vertical')
```
